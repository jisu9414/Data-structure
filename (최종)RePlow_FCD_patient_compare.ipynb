```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import MaxNLocator
```

```python
sample = input("")
patientA_count_over = int(input(""))
patientB_count_over = int(input(""))
patientC_count_over = int(input(""))
```

```python
patientA = pd.read_table(f'/content/patientA_{sample}.txt', sep='\t')
patientB = pd.read_table(f'/content/patientB_{sample}.txt', sep='\t')
patientC = pd.read_table(f'/content/patientC_{sample}.txt', sep='\t')
```

```python
df1 = patientA.iloc[:, [0, 6]]
df1_count = pd.DataFrame(df1['Gene.refGene'].value_counts(dropna=True, sort=False))
df1_gene = df1.drop_duplicates(subset='Gene.refGene')
df1_count.insert(0, 'chr', list(df1_gene['Chr']))
df1_count.columns = ['chr', 'count']
#df1_count['chr'] = df1_count['chr'].astype(str)

df1_count['gene_temp'] = df1_count.index
df1_count = df1_count.reset_index(drop=True)

df1_count.insert(1, 'gene', list(df1_count['gene_temp']))
del df1_count['gene_temp']

df1_count
```

```python
plt.rcParams["figure.figsize"] = (25, 5) 
plt.rcParams["font.size"] = 10
sns.set_style('whitegrid')

color = sns.color_palette("Paired")
pl1 = sns.barplot(x=df1_count['gene'], y = df1_count['count'], hue = df1_count['chr'], dodge=False, palette = color)
pl1.legend(bbox_to_anchor=(1, 1), ncol = 2, fontsize = 10, title = "chr", title_fontsize = 14)

plt.title(f'{sample}_patientA', fontsize = 15)
plt.xlabel('gene', fontsize=14)
plt.ylabel(f'snv count (>0)', fontsize=14)
pl1.set_xticklabels(pl1.get_xticklabels(), rotation=90)
#pl1.set_yticklabels(pl1.get_yticklabels(), rotation=90)
pl1.yaxis.set_major_locator(MaxNLocator(integer=True))
plt.show()
```

```python
df1_sort = df1_count.sort_values('count', ascending=False)
df1_sort = df1_sort.reset_index(drop=True)
df1_sort
```

```python
df1_clip = df1_count[df1_count['count'] > patientA_count_over]
df1_clip = df1_clip.reset_index(drop=True)
df1_clip
```

```python
plt.rcParams["figure.figsize"] = (25, 5) 
plt.rcParams["font.size"] = 10
sns.set_style('whitegrid')

color = sns.color_palette("Paired")
pl1 = sns.barplot(x=df1_clip['gene'], y = df1_clip['count'], hue = df1_clip['chr'], dodge=False, palette = color)
pl1.legend(bbox_to_anchor=(1, 1), ncol = 2, fontsize = 10, title = "chr", title_fontsize = 12)

plt.title(f'{sample}_patientA', fontsize = 15)
plt.xlabel('gene', fontsize=14)
plt.ylabel(f'snv count (>{patientA_count_over})', fontsize=14)
pl1.set_xticklabels(pl1.get_xticklabels(), rotation=90)
#pl1.set_yticklabels(pl1.get_yticklabels(), rotation=90)
pl1.yaxis.set_major_locator(MaxNLocator(integer=True))
plt.show()
```

```python
fig = plt.figure(figsize=(10, 10))
sns.set_style('whitegrid')

gene_name = []
for i in range(len(df1_clip['gene'])):
    if df1_clip['count'][i] > (patientA_count_over + 1):
        gene_name.append("chr" +  str(df1_clip['chr'][i]) + "\n" + str(df1_clip['gene'][i]))
    else:
        gene_name.append("")
gene_count=df1_clip['count']

separate = [0.05 for i in range(len(df1_clip.index))]

plt.pie(gene_count, labels=gene_name, textprops={'fontsize': 10}, autopct=lambda p : '{:.1f}%'.format(p) if p > 4 else "", startangle=90, explode=separate)
#plt.legend(gene_name, bbox_to_anchor=(1.2, 0.7), ncol = 3, fontsize = 10)

plt.title(f'{sample}: patientA', fontsize = 15)
plt.show()
```

```python
df3 = patientC.iloc[:, [0, 6]]
df3_count = pd.DataFrame(df3['Gene.refGene'].value_counts(dropna=True, sort=False))
df3_gene = df3.drop_duplicates(subset='Gene.refGene')
df3_count.insert(0, 'chr', list(df3_gene['Chr']))
df3_count.columns = ['chr', 'count']
df3_count['chr'] = df3_count['chr'].astype(str)

df3_count['gene_temp'] = df3_count.index
df3_count = df3_count.reset_index(drop=True)

df3_count.insert(1, 'gene', list(df3_count['gene_temp']))
del df3_count['gene_temp']

df3_count
```

```python
df3_sort = df3_count.sort_values('count', ascending=False)
df3_sort = df3_sort.reset_index(drop=True)
df3_sort
```

```python
df3_clip = df3_count[df3_count['count'] > patientC_count_over]
df3_clip = df3_clip.reset_index(drop=True)
df3_clip
```

```python
plt.rcParams["figure.figsize"] = (10, 5) 
plt.rcParams["font.size"] = 10
sns.set_style('whitegrid')

color = sns.color_palette("Paired")
pl3 = sns.barplot(x=df3_clip['gene'], y = df3_clip['count'], hue = df3_clip['chr'], dodge=False, palette = color)
pl3.legend(bbox_to_anchor=(1, 1), ncol = 1, fontsize = 14, title = "chr", title_fontsize = 12)

plt.title(f'{sample}_patientC', fontsize = 15)
plt.xlabel('gene', fontsize=14)
plt.ylabel(f'snv count (>{patientC_count_over})', fontsize=14)
pl3.set_xticklabels(pl3.get_xticklabels())
#pl3.set_yticklabels(pl3.get_yticklabels(), rotation=90)
pl3.yaxis.set_major_locator(MaxNLocator(integer=True))
plt.show()
```

```python
fig = plt.figure(figsize=(10, 10))

gene_name = []
for i in range(len(df3_clip['gene'])):
    if df3_clip['count'][i] > (patientC_count_over + 1):
        gene_name.append("chr" +  str(df3_clip['chr'][i]) + "\n" + str(df3_clip['gene'][i]))
    else:
        gene_name.append("")
gene_count=df3_clip['count']

separate = [0.02 for i in range(len(df3_clip.index))]

plt.pie(gene_count, labels=gene_name, textprops={'fontsize': 10},autopct=lambda p : '{:.1f}%'.format(p) if p > 4 else "", startangle=90, explode=separate)
#plt.legend(gene_name, bbox_to_anchor=(1.2, 0.7), ncol = 3, fontsize = 10)

plt.title(f'{sample}: patientC', fontsize = 15)
plt.show()
```

```python
merge_diff = pd.merge(df1_count, df2_count, on='gene', how='outer')
merge_diff_clip = pd.merge(merge_diff, df3_count, on='gene', how='outer')

merge_diff_clip = merge_diff_clip.dropna(thresh=4)
merge_diff_clip = merge_diff_clip.reset_index(drop=True)

merge_diff_clip = merge_diff_clip.rename(columns={'chr_x' : 'patientA_chr'})
merge_diff_clip = merge_diff_clip.rename(columns={'chr_y' : 'patientB_chr'})
merge_diff_clip = merge_diff_clip.rename(columns={'chr' : 'patientC_chr'})

merge_diff_clip = merge_diff_clip.rename(columns={'count_x' : f'{sample}_patientA'})
merge_diff_clip = merge_diff_clip.rename(columns={'count_y' : f'{sample}_patientB'})
merge_diff_clip = merge_diff_clip.rename(columns={'count' : f'{sample}_patientC'})

merge_diff_clip.insert(0, f'{sample}_gene', list(merge_diff_clip['gene']))
del merge_diff_clip['gene']

merge_diff_clip
```

```python
sns.set_style('darkgrid')
color = sns.color_palette("Set2")
ax = merge_diff_clip.plot(x= f'{sample}_gene', y=[f'{sample}_patientA', f'{sample}_patientB', f'{sample}_patientC'], kind="bar", figsize=(8, 7), color=color, width = 0.5)

plt.legend(bbox_to_anchor=(1, 1), ncol = 1, fontsize = 10, title = "chr", title_fontsize = 11)

plt.title(f'{sample}: comparing patients', fontsize = 15)
plt.xlabel('gene', fontsize=11)
plt.ylabel(f'snv count (A > {patientA_count_over}, B >  {patientB_count_over}, C > {patientC_count_over} )', fontsize=11)
ax.set_xticklabels(ax.get_xticklabels(), rotation=360)
#ax.set_yticklabels(ax.get_yticklabels(), rotation=90)
ax.yaxis.set_major_locator(MaxNLocator(integer=True))
plt.show()
```

```python
df1_count['count'] = df1_count['count'].astype(str)
df2_count['count'] = df2_count['count'].astype(str)
df3_count['count'] = df3_count['count'].astype(str)

merge_diff_all = pd.merge(df1_count, df2_count, on='gene', how='outer')
merge_diff_com = pd.merge(merge_diff_all, df3_count, on='gene', how='outer')

merge_diff_com = merge_diff_com.rename(columns={'chr_x' : 'patientA_chr'})
merge_diff_com = merge_diff_com.rename(columns={'chr_y' : 'patientB_chr'})
merge_diff_com = merge_diff_com.rename(columns={'chr' : 'patientC_chr'})

merge_diff_com = merge_diff_com.rename(columns={'count_x' : 'count'})
merge_diff_com = merge_diff_com.rename(columns={'count_y' : 'count'})
merge_diff_com = merge_diff_com.rename(columns={'count' : 'count'})

merge_diff_com.insert(0, f'{sample}_gene', list(merge_diff_com['gene']))
del merge_diff_com['gene']

merge_diff_com
```

```python
com_gene = merge_diff_com.dropna(axis=0)
com_gene.to_csv(f'{sample}_com_gene.txt', sep = '\t', index = False)
com_gene
```
