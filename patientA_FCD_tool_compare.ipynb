```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import MaxNLocator
```

```python
sample = input("")
```

```python
df1_Mutect2 = pd.read_table(f'/content/{sample}_Mutect2.txt', sep='\t')
df2_RePlow = pd.read_table(f'/content/{sample}_RePlow.txt', sep='\t')
df3_Strelka2 = pd.read_table(f'/content/{sample}_Strelka2.txt', sep='\t')
```

```python
plt.figure(figsize=(20, 10))

sns.swarmplot(data=df1_Mutect2, x = 'Chr', y = 'Start', color='black', size = 10, alpha = 0.3)
sns.swarmplot(data=df2_RePlow, x = 'Chr', y = 'Start', color='r', size = 12, alpha = 0.8)
sns.swarmplot(data=df3_Strelka2, x = 'Chr', y = 'Start', color='y', size = 8, alpha = 0.5)

#y축 눈금 수정
plt.tick_params(axis='x', direction='in', labelsize=14)
plt.tick_params(axis='y', direction='inout', labelsize=14)
current_values = plt.gca().get_yticks()
plt.gca().set_yticklabels(['{:,.0f}'.format(x) for x in current_values])

plt.show()
```

```python
df1_M = df1_Mutect2.iloc[:, [0, 6]]
df1_M_count = pd.DataFrame(df1_M['Gene.refGene'].value_counts(dropna=True, sort=False))
df1_M_gene = df1_M.drop_duplicates(subset='Gene.refGene')
df1_M_count.insert(0, 'chr', list(df1_M_gene['Chr']))
df1_M_count.columns = ['chr', 'count']
#df1_M_count['chr'] = df1_M_count['chr'].astype(str)

df1_M_count['gene_temp'] = df1_M_count.index
df1_M_count = df1_M_count.reset_index(drop=True)

df1_M_count.insert(1, 'gene', list(df1_M_count['gene_temp']))
del df1_M_count['gene_temp']
```

```python
df1_M_count
```

```python
plt.rcParams["figure.figsize"] = (20, 5) 
plt.rcParams["font.size"] = 10
sns.set_style('whitegrid')

color = sns.color_palette("Paired")
pl1 = sns.barplot(x=df1_M_count['gene'], y = df1_M_count['count'], hue = df1_M_count['chr'], dodge=False, palette = color)
pl1.legend(bbox_to_anchor=(1, 1), ncol = 2, fontsize = 10, title = "chr", title_fontsize = 14)

plt.title(f'Mutect2: {sample}', fontsize = 15)
plt.xlabel('gene', fontsize=14)
plt.ylabel(f'snv count (>0)', fontsize=14)
pl1.set_xticklabels(pl1.get_xticklabels())
#pl1.set_yticklabels(pl1.get_yticklabels(), rotation=90)
pl1.yaxis.set_major_locator(MaxNLocator(integer=True))
plt.show()
```

```python
fig = plt.figure(figsize=(10, 10))
sns.set_style('whitegrid')

gene_name = []
for i in range(len(df1_M_count['gene'])):
    if df1_M_count['count'][i] > 1:
        gene_name.append("chr" +  str(df1_M_count['chr'][i]) + "\n" + str(df1_M_count['gene'][i]))
    else:
        gene_name.append("")
gene_count=df1_M_count['count']

separate = [0.02 for i in range(len(df1_M_count.index))]

plt.pie(gene_count, labels=gene_name, textprops={'fontsize': 13}, autopct=lambda p : '{:.1f}%'.format(p) if p > 10 else "", startangle=90, explode=separate)
plt.legend(gene_name, bbox_to_anchor=(1.1, 0.9), ncol = 2, fontsize = 12)

plt.title(f'Mutect2: {sample}', fontsize = 15)
plt.show()
```

```python
df3_S = df3_Strelka2.iloc[:, [0, 6]]
df3_S_count = pd.DataFrame(df3_S['Gene.refGene'].value_counts(dropna=True, sort=False))
df3_S_gene = df3_S.drop_duplicates(subset='Gene.refGene')
df3_S_count.insert(0, 'chr', list(df3_S_gene['Chr']))
df3_S_count.columns = ['chr', 'count']
#df3_S_count['chr'] = df3_S_count['chr'].astype(str)

df3_S_count['gene_temp'] = df3_S_count.index
df3_S_count = df3_S_count.reset_index(drop=True)

df3_S_count.insert(1, 'gene', list(df3_S_count['gene_temp']))
del df3_S_count['gene_temp']

df3_S_count
```

```python
df1_M_count['count'] = df1_M_count['count'].astype(str)
df2_R_count['count'] = df2_R_count['count'].astype(str)
df3_S_count['count'] = df3_S_count['count'].astype(str)
```

```python
merge_diff_all = pd.merge(df1_M_count, df2_R_count, on='gene', how='outer')
merge_diff_com = pd.merge(merge_diff_all, df3_S_count, on='gene', how='outer')

merge_diff_com = merge_diff_com.rename(columns={'chr_x' : 'Mutect2_chr'})
merge_diff_com = merge_diff_com.rename(columns={'chr_y' : 'RePlow_chr'})
merge_diff_com = merge_diff_com.rename(columns={'chr' : 'Strelka2_chr'})

merge_diff_com = merge_diff_com.rename(columns={'count_x' : f'Mutect2: {sample}'})
merge_diff_com = merge_diff_com.rename(columns={'count_y' : f'RePlow: {sample}'})
merge_diff_com = merge_diff_com.rename(columns={'count' : f'Strelka2: {sample}'})

merge_diff_com.insert(0, f'{sample}_gene', list(merge_diff_com['gene']))
del merge_diff_com['gene']

merge_diff_com
```

```python
com_gene = merge_diff_com.dropna(axis=0)
com_gene.to_csv(f'{sample}_com_gene.txt', sep = '\t', index = False)
com_gene
```
